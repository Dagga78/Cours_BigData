```python
df.sum()
```

```python
sum = 0

for index, row in df.iterrows():
    row['item_price'] = row['item_price'].replace('$','')
    sum += float(row['item_price']) * float(row['quantity'])

print("Le chiffre d'affaires de Chipotle est de " + str(round(sum,2))  + "$")
    
nb_commande = np.count_nonzero(df.order_id.unique())

avg = sum / nb_commande
print("Le revenu moyen par commande est de " + str(round(avg,2)) + "$")
```
